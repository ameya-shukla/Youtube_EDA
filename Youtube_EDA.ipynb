***Question :***

Drop duplicates where `video_id` is getting duplicated for tseries data and **keep the last row** and then check how many rows are left.

```python
tseries["video_id"].duplicated().sum()
```

```python
tseries[tseries["video_id"].duplicated() == True]
```

```python
# drop the duplicates by looking at the subset ("video_id") but keep the "last" row of existence of the dupliacted 
# value which will keep the last or the most recent entry of that value

tseries.drop_duplicates(subset="video_id", keep="last", inplace=True)
```

```python
tseries.head()
```

```python
tseries.shape
```

```python
tseries["pub_date"] = pd.to_datetime(tseries["publish_date"])
```

```python
tseries["pub_day"] = tseries["pub_date"].dt.day
```

```python
tseries["pub_month"] = tseries["pub_date"].dt.month
```

```python
tseries["pub_year"] = tseries["pub_date"].dt.year
```

```python
tseries
```

```python
import numpy as np
plt.title("Monthly frequency of Video Uploads of T-Series")
sns.histplot(tseries["pub_month"], bins=12)
plt.xticks(np.arange(1,13))
plt.xlabel("Video Publish Month")
plt.show()
```

```python
tseries.pub_month.value_counts()
```

***Insights :***

* No video uploads in months `7`,`8`,`9`,`10`
    * Possible Reasons:
        * Exam period in India in months `7`,`8`,`9`,`10` which can lead to low views
  
  
* Highest uploads : 
    * Possible Reasons:
        * month `2` (Feb) : *Valentines day*
        * month `3` (Mar) : *Holi*
        * month `12` (Dec) and `1` (Jan) : *Wedding season*
